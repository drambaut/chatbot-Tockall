# Instalación de modelo de conversión a embbedings

```python
# Importar modelo de transformacion a embedings
from torch import cuda
from langchain.embeddings.huggingface import HuggingFaceEmbeddings

embed_model_id = 'sentence-transformers/all-MiniLM-L6-v2'

device = f'cuda:{cuda.current_device()}' if cuda.is_available() else 'cpu'

embed_model = HuggingFaceEmbeddings(
    model_name=embed_model_id,
    model_kwargs={'device': device},
    encode_kwargs={'device': device, 'batch_size': 32}
)
```

```python
# Prueba de funcionamiento de embedings
docs = [
    "this is one document",
    "and another document"
]

embeddings = embed_model.embed_documents(docs)

print(f"We have {len(embeddings)} doc embeddings, each with "
      f"a dimensionality of {len(embeddings[0])}.")
```

```python
import os
from pinecone import Pinecone

api_key = os.environ.get('PINECONE_API_KEY') or 'be227b3a-d248-411d-8808-00b17f911811'

pc = Pinecone(api_key=api_key)
```

# Importe de documentación

```python
# Importe del dataframe
from datasets import load_dataset
import pandas as pd
df=pd.read_excel("reglamento.xlsx")
data = df
data
```

```python
# Personalización del dataframe para tomar los datos que se necesitan
batch_size = 32

for i in range(0, len(data), batch_size):
    i_end = min(len(data), i+batch_size)
    batch = data.iloc[i:i_end]
    ids = [f"{x['doi']}-{x['chunk-id']}" for i, x in batch.iterrows()]
    texts = [x['chunk'] for i, x in batch.iterrows()]
    embeds = embed_model.embed_documents(texts)
    metadata = [
        {'text': x['chunk'],
         'source': x['source'],
         'title': x['title']} for i, x in batch.iterrows()
    ]
    index.upsert(vectors=zip(ids, embeds, metadata))
```

# Instalación y aplicación de la IA llama2 al código

```python
from torch import cuda, bfloat16
import transformers

model_id = 'meta-llama/Llama-2-7b-chat-hf'

device = f'cuda:{cuda.current_device()}' if cuda.is_available() else 'cpu'


bnb_config = transformers.BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type='nf4',
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=bfloat16
)

hf_auth = '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'
model_config = transformers.AutoConfig.from_pretrained(
    model_id,
    use_auth_token=hf_auth
)

model = transformers.AutoModelForCausalLM.from_pretrained(
    model_id,
    trust_remote_code=True,
    config=model_config,
    quantization_config=bnb_config,
    device_map='auto',
    use_auth_token=hf_auth
)
model.eval()
print(f"Model loaded on {device}")
```

```python
tokenizer = transformers.AutoTokenizer.from_pretrained(
    model_id,
    use_auth_token=hf_auth
)
```

```python
# Pipeline para el funcionamiento de la generacion de respuestas por parte de la ia
generate_text = transformers.pipeline(
    model=model, tokenizer=tokenizer,
    return_full_text=True,  # langchain expects the full text
    task='text-generation',
    temperature=0.0,
    max_new_tokens=512,
    repetition_penalty=1.1
)
```

```python
# Prueba de funcionamiento de respuesta de ia
res = generate_text("A cuantos pesos equivale el dolar hoy")
print(res[0]["generated_text"])
```

```python
# Pipeline para facilitar la manera de preguntarle a la ia

from langchain.llms import HuggingFacePipeline

llm = HuggingFacePipeline(pipeline=generate_text)
```

```python
from langchain_pinecone import PineconeVectorStore

text_field = 'text'

vectorstore = PineconeVectorStore(index, embed_model.embed_query, text_field)
```

```python
# Prueba de funcionamiento de funcion de busqueda por similaridad

query = 'cual es el horario de atencion del restaurante casa san isidro'

vectorstore.similarity_search_by_vector_with_score(
            embed_model.embed_query(query), k=5
            )
```

# Creacion de cadena para generar la respuesta de acuerdo con modelo RAG

```python
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate

# Definir la consulta y vectorizarla
query = 'cual es el menu de carnes del restaurante casa san isidro'
query_vector = embed_model.embed_query(query)

# Realizar la búsqueda de similitud
results = vectorstore.similarity_search_by_vector_with_score(query_vector, k=3)

# Crear una plantilla de prompt para la cadena del LLM
template = """You are an AI assistant. You have access to the following information:

{text}

Based on this information, provide a comprehensive answer to the following question:
{question}
"""
prompt = PromptTemplate(template=template, input_variables=["text", "question"])

# Definir la cadena del LLM
chain = LLMChain(
    llm=llm,
    prompt=prompt
)

# Utilizar la cadena con los resultados de búsqueda
context = " ".join([result[0].page_content for result in results])
question = "cual es el menu de carnes del restaurante casa san isidro"
response = chain.run({"text": context, "question": question})

print(response)
```
